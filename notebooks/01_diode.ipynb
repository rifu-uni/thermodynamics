```python
import pandas as pd, matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/01_diode/diode.csv', header=0)
data
```

```python
fig = plt.figure(figsize=(15, 10))

# Upper left: Polar
ax1 = fig.add_subplot(2, 2, 1)
ax1.scatter(data['temp'], data['voltLM35'], linewidth=1.2, label='LM35', color='xkcd:blue')
ax1.set_title('LM35', fontsize=16, fontweight='bold', pad=20)
ax1.set_ylabel('voltage', fontsize=14)
ax1.set_xlabel('temp', fontsize=14)
ax1.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)

# Upper right: Polar
ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter(data['temp'], data['voltDiodo'], linewidth=1.2, label='diodo', color='xkcd:teal')
ax2.set_title('Diode', fontsize=16, fontweight='bold', pad=20)
ax2.set_xlabel('temp', fontsize=14)
ax2.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)
```
