```python
## Calorimetro
k = (121*(88.7-46.2)-156.1*(46.2-18.7))/(46.2-18.7)
print(f'Equivalente en agua del calorimetro: {k:.3f}\n')

def calor(ma, Tf, Ta0, mx, Tx0, ve, k=k, ca=1):
    res = ((ma+k)*ca*(Tf - Ta0))/(mx*(Tx0 - Tf))
    err = 100*(res-ve)/ve

    return res, err


res, err = calor(231.2, 21.5, 19.7, 53.1, 90.7, 0.113)
print(rf"Hierro : {res:.4f}, {err:.3f}%")

res, err = calor(116.1, 22.4, 20.3, 18.7, 90.7, 0.217)
print(rf"Aluminio: {res:.4f}, {err:.3f}%")

res, err = calor(119.1, 23.6, 21.9, 33.7, 90.4, 0.092)
print(rf"Bronce: {res:.4f}, {err:.3f}%")
```
