```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = []
data = []
data1 = []
with open('data/pt100Thermistor/mixedHeating1.csv', mode='r', newline='') as file:
    reader = csv.reader(file)
    header = next(reader)
    for row in reader:
        time.append(float(row[0]))
        data.append(float(row[1]))
        data1.append(float(row[6]))
```

```python
plt.style.use('seaborn-v0_8-whitegrid')  # Modern style
fig, ax = plt.subplots(figsize=(16, 8))

ax.plot(time, data, marker='.', linewidth=2, color='xkcd:teal', label='PT100', zorder=1)
# ax.plot(time, data1, marker='.', linewidth=2, color='xkcd:orange', label='Thermistor', zorder=1)
ax.set_xlabel(r'$t \, [s]$', fontsize=14)
ax.set_ylabel(r'$v$', fontsize=14)
ax.set_title('Res', fontsize=16, fontweight='bold', pad=15)
# ax.tick_params(axis='both', which='major', labelsize=10)
# ax.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.7)
ax.legend(fontsize=12, loc='best', frameon=True)
# plt.tight_layout()
# plt.savefig('Srel.png', dpi=300)
plt.show()
```
